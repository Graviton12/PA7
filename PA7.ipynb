* Build a Scientific calculator which will have at least 15 scientific operations. You can use basic statistics, and math classes

```python
import math 

class calculator1():
    def __init__(self, input1, input2):
        self.input1=input1
        self.input2=input2
    def addition(self):
        return self.input1 + self.input2
    def subtraction(self):
        return self.input1 - self.input2  
    def multiplication(self):
        return self.input1 * self.input2
    def division(self):
        return self.input1 / self.input2
class calculator2():
    def __init__(self, only_input1):
        self.only_input1=only_input1
    def factorial(self):
        return math.factorial(self.only_input1)
    def square(self):
        return math.pow(self.only_input1, 2) 
    def cube(self):
        return math.pow(self.only_input1, 3)
    def logarithm(self):
        return math.log(self.only_input1)
    def e_to_the_x(self):
        return math.exp(self.only_input1)   
    def square_root(self):
        return math.sqrt(self.only_input1)    
    def absolute_value(self):
        return math.fabs(self.only_input1)
    def cos(self):
        return math.cos(self.only_input1)
    def sin(self):
        return math.sin(self.only_input1)
    def tan(self):
        return math.tan(self.only_input1)
    def rad_to_deg(self):
        return math.degrees(self.only_input1)
    def deg_to_rad(self):
        return math.radians(self.only_input1)

calc_choice = int(input("Select 1. for basic math opperations (two inputs) and 2. for advanced math opperations (one input): "))
if calc_choice==1:
    input_1=int(input("Enter first number: "))
    input_2=int(input("Enter second number: "))
    my_calc1=calculator1(input_1, input_2)
    print()   
    print("0. Exit")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division") 
    
    choice=1
    while choice!=0:
        choice=int(input("Enter your arithmetic choice... "))
        if choice==1:
            print(f"\tResult: {my_calc1.addition()}")
        elif choice==2:
            print(f"\tResult: {my_calc1.subtraction()}")
        elif choice==3:
            print(f"\tResult: {my_calc1.multiplication()}")
        elif choice==4:
            print(f"\tResult: {my_calc1.division()}")
        elif choice==0:
            print("Exit")
        else:
            print("Sorry, invalid choice")
    print()   
elif calc_choice==2:
    input_1=int(input("Enter a number: "))
    my_calc2=calculator2(input_1)
    print()
    print("0. Exit")    
    print("5. Factorial")
    print("6. Square")
    print("7. Cube")
    print("8. Logarithm")
    print("9. e^x")
    print("10. Square Root")
    print("11. Absolute Value")
    print("12. Cosine")
    print("13. Sine")
    print("14. Tangent")
    print("15. Radian to Degree")
    print("16. Degree to Radian")
    
    choice=1
    while choice!=0:
        choice=int(input("Enter your arithmetic choice... "))
        if choice==5:
            print(f"\tResult: {my_calc2.factorial()}")
        elif choice==6:
            print(f"\tResult: {my_calc2.square()}")
        elif choice==7:
            print(f"\tResult: {my_calc2.cube()}")    
        elif choice==8:
            print(f"\tResult: {my_calc2.logarithm():.4f}") 
        elif choice==9:
            print(f"\tResult: {my_calc2.e_to_the_x():.4f}") 
        elif choice==10:
            print(f"\tResult: {my_calc2.square_root():.4f}") 
        elif choice==11:
            print(f"\tResult: {my_calc2.absolute_value()}")
        elif choice==12:
            print(f"\tResult: {my_calc2.cos():.4f}")
        elif choice==13:
            print(f"\tResult: {my_calc2.sin():.4f}")
        elif choice==14:
            print(f"\tResult: {my_calc2.tan():.4f}")
        elif choice==15:
            print(f"\tResult: {my_calc2.rad_to_deg():.4f}") 
        elif choice==16:
            print(f"\tResult: {my_calc2.deg_to_rad():.4f}")    
        elif choice==0:
            print("Exit")
else:
    print("Sorry, invalid choice. Select again")
```
